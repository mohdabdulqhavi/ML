1. What is the concept of supervised learning? What is the significance of the name?
* Supervised learning is a type of machine learning where the model is trained on labeled data to predict outcomes for new, unseen data. The term "supervised" refers to the fact that the model is trained with labeled examples and supervised during the learning process.

4. In supervised learning, what are classification and regression?
* Classification is a type of supervised learning where the model is trained to predict the class label of a given input data. Regression is another type of supervised learning where the model is trained to predict a continuous output value based on the input data.

6. Briefly describe the SVM model.
* Support Vector Machines (SVMs) are supervised learning models used for classification and regression problems. It works by mapping data to a high-dimensional feature space and constructing a hyperplane or set of hyperplanes in the space that best divides the classes. It is an effective method for both linear and non-linear classification and regression.

--

7. In SVM, what is the cost of misclassification?
* The cost of misclassification in SVM is the misclassification error rate, which is the percentage of training examples that are incorrectly classified. This error rate is used to determine the penalty for misclassification and is an important factor in the optimization of the SVM model.

--

8. In the SVM model, define Support Vectors.
* Support vectors are the data points that lie closest to the decision boundary of the SVM model. They are the data points that define the maximum margin hyperplane between the two classes in a multi-dimensional space. In addition, they are the most influential data points in the model.

--

9. In the SVM model, define the kernel.
* Kernel is a mathematical function used to map a lower dimensional data into a higher dimensional space. It can be used for various algorithms such as SVM, K Nearest Neighbour, Logistic Regression etc. In SVM, kernel is used to transform the data into a higher dimensional space to separate the data using a hyperplane. The most common types of kernels are linear, polynomial, radial basis function (RBF) and sigmoid. Each of these kernels has its own advantages and disadvantages and is used for different scenarios.

--

10. What are the factors that influence SVM&#39;s effectiveness?
* The factors that influence SVM's effectiveness include the size of the dataset, kernel selection, regularization parameters, choice of hyperparameters, and data preprocessing techniques. Additionally, factors such as feature scaling, cost, and kernel parameters can also affect the accuracy, speed, and performance of SVM.

11. What are the benefits of using the SVM model?

It has the ability to efficiently handle high dimensional data, and perform well even with small datasets. It also provides good generalization performance and has a powerful kernel trick for non-linear data. 16. Explain the decision tree algorithm in a few words.
* Decision tree algorithm is a supervised learning technique that uses a tree-like structure to identify a decision path based on known input data.

--

17. What is the difference between a node and a leaf in a decision tree?
* A node is an internal point in a decision tree that has two or more branches extending from it, while a leaf is a terminal point in a decision tree that has no branches extending from it. The nodes are used to split the dataset into smaller subsets based on certain criteria, while the leaves are used to make a prediction or classification.

--

18. What is a decision tree&#39;s entropy?
* Decision tree entropy is a measure of the randomness or disorder in the data set. It is calculated based on the probabilities of each outcome in the data set. The higher the entropy, the less ordered the data set is and the more difficult it is to make accurate predictions.

--

19. In a decision tree, define knowledge gain.
* Knowledge gain is a measure of how much information is gained by splitting a node in a decision tree. It is calculated by subtracting the entropy of the parent node from the weighted average entropy of the two child nodes. The higher the knowledge gain, the more effective the split is in reducing the uncertainty in the data.

22. Briefly describe the random forest model.
* Random forest is an ensemble learning method for both classification and regression. It is a type of decision tree algorithm which works by constructing a multitude of decision trees at training time and outputting the class that is the mode of the classes (classification) or mean prediction (regression) of the individual trees. It is a powerful and popular algorithm due to its ability to handle a large number of input variables, create accurate models, and reduce overfitting. Random forest models are also highly interpretable and relatively easy to use.
